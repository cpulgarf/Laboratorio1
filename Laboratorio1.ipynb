Elabore las siguientes funciones:
    
**1. La función `describir(datos)` que genere un arreglo con strings que describan cada fruta de la siguiente manera:**

```{El/La} {...} es una fruta de tipo {...} de color {...}. {Presenta/No presenta} pepas en su interior.```
    
Por ejemplo, para el plátano se debe generar el siguiente string:


**2. La función `filtrar_por_pepa(datos, tiene_pepas)` que dado un tipo de fruta y un booleano, retorne las frutas que tienen o no tienen pepas (según el valor de `tiene_pepas`).**
    

Por ejemplo, `filtrar_por_pepa(datos, True)` deberá retornar un arreglo con los diccionarios de: naranja, limon, frutilla y frambuesa.


**3. La función `conteo_colores(datos)` que cree un diccionario que haga un conteo los colores de las frutas.**

Para estos datos, la función debería retornar el siguiente diccionario: 
    
```python
{"amarillo": 3, "naranjo": 1, "rojo": 2}
````

**Desarrollo de código:**

`````python
def describir(datos):
    # Código Aquí
    texto = []
    for i in datos:
      
      nombre = i["nombre"]
      tipo = i["tipo"]
      color = i["color"]
      pepas = i["pepas"]

      if nombre[-1] == "a":
        if pepas == True:
          te = "La" +" " + nombre + " " + "es una fruta de tipo" + " " + tipo + " " + "de color" + " " + color + ". Presenta pepas en su interior." 
          texto.append(te)
        else:
          te = "La" +" " + nombre + " " + "es una fruta de tipo" + " " + tipo + " " + "de color" + " " + color + ". No presenta pepas en su interior." 
          texto.append(te)
      else:
        if pepas == True:
         te = "El" + " " + nombre + " " + "es una fruta de tipo"+ " " + tipo + " " + "de color"+ " " + color + ". Presenta pepas en su interior."
         texto.append(te)
        else:
          te = "El" + " " + nombre + " " + "es una fruta de tipo"+ " " + tipo + " " + "de color"+ " " + color + ". No presenta pepas en su interior."
          texto.append(te) 
    return texto


def filtrar_por_pepa(datos, tiene_pepas):
    # Código Aquí
    filtrados = []
    for i in datos:
      pepas = i["pepas"]
      if pepas == tiene_pepas:
        filtrados.append(i)
    return filtrados



def conteo_colores(datos):
    # Código Aquí
    conteo= {}
    for fruta in datos:
      color = fruta["color"]
      if color not in conteo.keys():
        conteo[color] = 1
      else:
        conteo[color] += 1
    return conteo
`````

## Bonus [0.5]

Implementar `bandeja` usando una `property` que permita setear una `nueva_bandeja` como `bandeja` según las siguientes condiciones:


- Compruebe que `nueva_bandeja` sea una lista. En caso contrario, levante una excepción.
- No permita agregar más de 3 ingredientes a la bandeja a la vez. Si se entregan más de 3 frutas, se levante una excepción.
- Se compruebe que todos los elementos del arreglo sean frutas. Para esto, por cada fruta compruebe que:
    1. La fruta sea diccionario.
    2. El diccionario entregado tenga las llaves `nombre`, `tipo`, `color` y `pepas`.

`````python
import copy
class Jugera(Electrodomestico):
    
    
    def __init__(self):
      super().__init__()
      # Código Aquí
      self._bandeja = []

    @property
    def bandeja(self):
        # Código Aquí
        return self._bandeja
        
    @bandeja.setter
    def bandeja(self, nueva_bandeja):
        # Código Aquí       
        if type(nueva_bandeja) == list:   #Verifica si la bandeja está en forma de lista
          if len(nueva_bandeja) < 4:      #verifica si hay máximo 3 ingredientes en la jugera
            for elemento in nueva_bandeja:   #Verificar por elemento dentro de la lista (frutas en la badeja) son diccionarios y tienen las llaves: nombre, tipo, color y pepas.
              llaves_frutas = list(elemento.keys())
              llaves_orig = ['nombre', 'tipo', 'color', 'pepas']
              if type(elemento) == dict and set(llaves_frutas) == set(llaves_orig):   
                continue
              
              else:
                raise Exception('Cuidado ⚠️: El formato de las frutas no es diccionario o no tiene la información necesaria')
            self._bandeja = nueva_bandeja
          else:
            raise Exception('Cuidado ⚠️: ¡La jugera está llena!.')
        else:
          raise Exception('Cuidado ⚠️: La bandeja debe ser una lista!.')


    def agregar_ingrediente(self, nueva_fruta):
        # Código Aquí
        band =  self._bandeja
        if nueva_fruta not in band:
          if type(nueva_fruta)== dict:
            band.append(nueva_fruta)
          else:
            raise Exception('Cuidado ⚠️: Debes añadir una nueva fruta como diccionario.')
        else:
          raise Exception('Cuidado ⚠️: ¡Ya está la fruta en la jugera!.')

    def listar_ingredientes(self):
        # Código Aquí
        if self._bandeja == []:
          print("Bandeja vacía")
        else:
          frutas = []
          for fruta in self._bandeja:
            frutas.append(fruta["nombre"])
          print("Ingredientes en la bandeja: " + ', '.join(frutas))

    
    def preparar_jugo(self):
        # Código Aquí
        if self.esta_enchufado() == True:
          if len(self._bandeja) > 0:
            if len(filtrar_por_pepa_funcional(self._bandeja, True)) == 0:  #uso de la funcion "filtrar_por_pepa_funcional" usando los datos de "bandeja", en 
              ba = self._bandeja                             #el caso de que el largo de la lista sea mayor a 0, significa que hay almenos una fruta con pepas
              frut = []
              for fruta in self._bandeja:
                  frut.append(fruta["nombre"])
              ba.clear()  #vaciar la jugera caundo se hace el jugo
              print("Jugo de: " + ', '.join(frut)+ " listo. 🏖️🥤 Que lo disfrutes!!! 🥤🏖️.")
            
            else:
              ba = self._bandeja  
              ba.clear()  #vaciar la jugera cuando hay una fruta con pepas
              raise Exception('Alerta ⚠️: El jugo puede contener restos de pepas.')
                           
          else:
            raise Exception('Error ❌: La bandeja no tiene ingredientes.') 
`````
