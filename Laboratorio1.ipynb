`````python
class Electrodomestico:
    
    def __init__(self):
      # C√≥digo Aqu√≠
      self.enchufado = False
        
    def esta_enchufado(self):
      # C√≥digo Aqu√≠
      if self.enchufado == False:
         raise Exception('Error ‚ùå: El electrodom√©stico no est√° enchufado') 
      else:
        return True
            
    def enchufar(self):
      # C√≥digo Aqu√≠
      self.enchufado = True
`````

## Bonus [0.5]

Implementar `bandeja` usando una `property` que permita setear una `nueva_bandeja` como `bandeja` seg√∫n las siguientes condiciones:


- Compruebe que `nueva_bandeja` sea una lista. En caso contrario, levante una excepci√≥n.
- No permita agregar m√°s de 3 ingredientes a la bandeja a la vez. Si se entregan m√°s de 3 frutas, se levante una excepci√≥n.
- Se compruebe que todos los elementos del arreglo sean frutas. Para esto, por cada fruta compruebe que:
    1. La fruta sea diccionario.
    2. El diccionario entregado tenga las llaves `nombre`, `tipo`, `color` y `pepas`.

`````python
import copy
class Jugera(Electrodomestico):
    
    
    def __init__(self):
      super().__init__()
      # C√≥digo Aqu√≠
      self._bandeja = []

    @property
    def bandeja(self):
        # C√≥digo Aqu√≠
        return self._bandeja
        
    @bandeja.setter
    def bandeja(self, nueva_bandeja):
        # C√≥digo Aqu√≠       
        if type(nueva_bandeja) == list:   #Verifica si la bandeja est√° en forma de lista
          if len(nueva_bandeja) < 4:      #verifica si hay m√°ximo 3 ingredientes en la jugera
            for elemento in nueva_bandeja:   #Verificar por elemento dentro de la lista (frutas en la badeja) son diccionarios y tienen las llaves: nombre, tipo, color y pepas.
              llaves_frutas = list(elemento.keys())
              llaves_orig = ['nombre', 'tipo', 'color', 'pepas']
              if type(elemento) == dict and set(llaves_frutas) == set(llaves_orig):   
                continue
              
              else:
                raise Exception('Cuidado ‚ö†Ô∏è: El formato de las frutas no es diccionario o no tiene la informaci√≥n necesaria')
            self._bandeja = nueva_bandeja
          else:
            raise Exception('Cuidado ‚ö†Ô∏è: ¬°La jugera est√° llena!.')
        else:
          raise Exception('Cuidado ‚ö†Ô∏è: La bandeja debe ser una lista!.')


    def agregar_ingrediente(self, nueva_fruta):
        # C√≥digo Aqu√≠
        band =  self._bandeja
        if nueva_fruta not in band:
          if type(nueva_fruta)== dict:
            band.append(nueva_fruta)
          else:
            raise Exception('Cuidado ‚ö†Ô∏è: Debes a√±adir una nueva fruta como diccionario.')
        else:
          raise Exception('Cuidado ‚ö†Ô∏è: ¬°Ya est√° la fruta en la jugera!.')

    def listar_ingredientes(self):
        # C√≥digo Aqu√≠
        if self._bandeja == []:
          print("Bandeja vac√≠a")
        else:
          frutas = []
          for fruta in self._bandeja:
            frutas.append(fruta["nombre"])
          print("Ingredientes en la bandeja: " + ', '.join(frutas))

    
    def preparar_jugo(self):
        # C√≥digo Aqu√≠
        if self.esta_enchufado() == True:
          if len(self._bandeja) > 0:
            if len(filtrar_por_pepa_funcional(self._bandeja, True)) == 0:  #uso de la funcion "filtrar_por_pepa_funcional" usando los datos de "bandeja", en 
              ba = self._bandeja                             #el caso de que el largo de la lista sea mayor a 0, significa que hay almenos una fruta con pepas
              frut = []
              for fruta in self._bandeja:
                  frut.append(fruta["nombre"])
              ba.clear()  #vaciar la jugera caundo se hace el jugo
              print("Jugo de: " + ', '.join(frut)+ " listo. üèñÔ∏èü•§ Que lo disfrutes!!! ü•§üèñÔ∏è.")
            
            else:
              ba = self._bandeja  
              ba.clear()  #vaciar la jugera cuando hay una fruta con pepas
              raise Exception('Alerta ‚ö†Ô∏è: El jugo puede contener restos de pepas.')
                           
          else:
            raise Exception('Error ‚ùå: La bandeja no tiene ingredientes.') 
`````
